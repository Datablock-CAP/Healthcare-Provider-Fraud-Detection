```python
import pandas as pd
import numpy as np
```

## `Reading Datasets` 

```python
# inpatient dataset
inpatient = pd.read_csv('./../../archive (2)/Train_Inpatientdata-1542865627584.csv')
```

```python
# Outpatient dataset
outpatient = pd.read_csv('./../../archive (2)/Train_Outpatientdata-1542865627584.csv')
```

```python
inpatient.head()
```

```python
outpatient.head()
```

## `ClmAdmitDiagnosisCode`

### - merge the inpatient and outpatient datasets

```python
# concat the inpatient and outpatient datasets
in_out_patients = pd.concat([inpatient, outpatient])
```

```python
in_out_patients.head()
```

```python
# check if the number of samples are the same
in_out_patients.shape[0] == (inpatient.shape[0] + outpatient.shape[0])
```

## `Diagnosis Codes`

>To deal with missing values in the Diagnosis Code columns, we will replace those features with one feature which represents the number of diagnoses the patient has taken.

### - get diagnosis columns

```python
# define diagnosis columns
diagnosis_columns = list(inpatient.columns[14:24])
```

```python
diagnosis_columns
```

```python
# get the diagnosis columns
diagnosis_codes = in_out_patients[diagnosis_columns]
```

### - get the DiagnosisNumber column


```python
# replace non missing values with one
diagnosis_codes[~diagnosis_codes.isna()] = 1
```

```python
# replace missing values with zero
diagnosis_codes[diagnosis_codes.isna()] = 0
```

```python
diagnosis_codes.head()
```

```python
# create the sum column
diagnosis_number = diagnosis_codes.sum(axis=1)
```

```python
diagnosis_number
```

```python
# add the column to in out patient dataframe
in_out_patients['DiagnosisNumber'] = diagnosis_number
```

```python
in_out_patients.head()
```

```python
# delete the diagnosis columns
in_out_patients = in_out_patients.drop(diagnosis_columns, 1)
```

```python
in_out_patients.columns
```
